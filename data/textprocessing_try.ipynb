```python
reviews=open('myOutFile2.txt').read()
```

```python
from nltk.corpus import stopwords 
from nltk.stem.wordnet import WordNetLemmatizer
import string
```

```python
stop = set(stopwords.words('english'))
exclude = set(string.punctuation) 
lemma = WordNetLemmatizer()
```

```python
def clean(doc):
    stop_free = " ".join([i for i in doc.lower().split() if i not in stop])
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())
    return normalized
```

```python
reviews = reviews.split('\n')
```

```python
reviews_cln= [clean(line).split() for line in reviews]   
```

```python
import gensim
from gensim import corpora
```

```python
dictionary = corpora.Dictionary(reviews_cln)
```

```python
doc_term_matrix = [dictionary.doc2bow(line) for line in reviews_cln]
```

```python
Lda = gensim.models.ldamodel.LdaModel
```

```python
ldamodel = Lda(doc_term_matrix, num_topics=3, id2word = dictionary, passes=50)
```

```python
print(ldamodel.print_topics(num_topics=3, num_words=3))
```

```python
ldamodel = Lda(doc_term_matrix, num_topics=10, id2word = dictionary, passes=10)
```

```python
print(ldamodel.print_topics(num_topics=10, num_words=10))
```
